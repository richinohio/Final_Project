```python
import os 
import pandas as pd
```

```python
filepath = os.path.join("Tenant_History.csv")
df = pd.read_csv(filepath, skiprows=1)
df.head(100)
```

```python
move_in_df = df[df["Event"] == "Move In"]
move_out_df = df[df["Event"] == "Move Out"]
```

```python
csv_df = pd.DataFrame(move_in_df.groupby(["Unit_Code", "Tenant_Code"]).count())
csv_df.reset_index(inplace=True)
csv_df.head()
csv_df.to_csv("unit_tenant_move_in.csv")
```
