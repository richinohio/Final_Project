```python
import pandas as pd 
import matplotlib.pyplot as plt
import numpy as np
from datetime import datetime as dt
```

```python
tenants = "Tenant_History.csv"
tenants_df = pd.read_csv(tenants)
tenants_df.head()
```

```python
tenants_df = tenants_df.drop(columns = ["Ref_Property_ID", "Ext_Ref_Tenant_Id", "Ref_Unit_Id", "Date"])
```

```python
tenants_df["DateOccurred"] = pd.to_datetime(tenants_df["DateOccurred"], format= '%m/%d/%Y')
tenants_df["DateApply"] = pd.to_datetime(tenants_df["DateApply"], format= '%m/%d/%Y')
tenants_df["DateApprove"] = pd.to_datetime(tenants_df["DateApprove"], format= '%m/%d/%Y')
tenants_df["DateSign"] = pd.to_datetime(tenants_df["DateSign"], format= '%m/%d/%Y')
tenants_df["DateMoveIn"] = pd.to_datetime(tenants_df["DateMoveIn"], format= '%m/%d/%Y')
tenants_df["DateMoveOut"] = pd.to_datetime(tenants_df["DateMoveOut"], format= '%m/%d/%Y')
tenants_df["DateLeaseFrom"] = pd.to_datetime(tenants_df["DateLeaseFrom"], format= '%m/%d/%Y')
tenants_df["DateLeaseTo"] = pd.to_datetime(tenants_df["DateLeaseTo"], format= '%m/%d/%Y')
tenants_df["DateNotice"] = pd.to_datetime(tenants_df["DateNotice"], format= '%m/%d/%Y')
tenants_df["Length of Stay"]= tenants_df["DateMoveOut"] - tenants_df["DateMoveIn"]
tenants_df = tenants_df.drop(columns = ["DateApply", "DateApprove", "DateSign", "DateLeaseFrom", "DateLeaseTo", "DateNotice", "RentedEvent", "UnRentedEvent", "Status", "Deposit", "Notes", "Reason", "MTM"])
tenants_df.sort_values(by=["Tenant_Code"])
tenants_df.columns
```

```python
def vacancy (row):
    if row["Event"] == "Move Out":
        return 1
    elif row["Event"] == "Move In":
        return 0
    else:
        return 0
    
tenants_df ["Vacancy"] = tenants_df.apply(vacancy, axis = 1) 
tenants_df ["Vacancy"] = tenants_df.Vacancy.astype(int)
tenants_df
```

```python
rent_change_tenants = tenants_df.loc[tenants_df["Event"].isin(["Rent Change"])]
rent_change_tenants.reset_index(drop = True)
```

```python
rentchange = pd.concat([tenants_df, rent_change_tenants]).drop_duplicates().reset_index(drop=True)
rentchange = rentchange.loc[rentchange["Event"].isin(["Rent Change", "Move Out"])]
rentchange.reset_index(drop = True)
rentchange = rentchange.sort_values(by = ["Tenant_Code", "DateOccurred"])
rentchange
```

```python
rentchange.to_csv("Rent_Changed_Tenants.csv")
```

```python
print("Total Unique Tenants:" + tenants_df["Tenant_Code"].nunique())
print("Tenants with Rent Change"  rentchange["Tenant_Code"].nunique())
print("Tenants without Rent Change"  (tenants_df["Tenant_Code"].nunique()) - (rentchange["Tenant_Code"].nunique()))
```
